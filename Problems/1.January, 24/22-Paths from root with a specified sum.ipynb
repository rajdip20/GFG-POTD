Given a Binary tree and a sum **S**, print all the paths, **starting from root**, that sums upto the given sum. Path **not necessarily** end on a leaf node.

<br>

**Example 1:**

>Input : <br>
>sum = 8<br>
>Input tree<br>
>&emsp;&emsp;&emsp;&emsp;1 <br>
>&emsp;&emsp;&emsp;&ensp;/&ensp;\ <br>
>&emsp;&emsp;&ensp;20&emsp;&ensp;3 <br>
>&emsp;&emsp;&emsp;&emsp;&ensp;/&emsp;&ensp;\ <br>
>&emsp;&emsp;&emsp;&emsp;4&emsp;&emsp;&ensp;15<br>
>&emsp;&emsp;&emsp;&ensp;/ &ensp;\ &emsp;&emsp;/ &ensp;\ <br>
>&emsp;&emsp;&ensp;6&emsp;&emsp;7&emsp;8&emsp;&emsp;9<br>
>**Output :**<br>
>1 3 4<br>
>**Explanation :** <br>
>Sum of path 1, 3, 4 = 8.


**Example 2:**

>**Input :**<br> 
>sum = 38<br>
>Input tree<br>
>&emsp;&emsp;&emsp;&emsp;10 <br>
>&emsp;&emsp;&emsp;&ensp;/&emsp;&ensp;\ <br>
>&emsp;&emsp;&ensp;28&emsp;&emsp;13 <br>
>&emsp;&emsp;&emsp;&emsp;&emsp;/&emsp;&ensp;\ <br>
>&emsp;&emsp;&emsp;&emsp;14&emsp;&emsp;&ensp;15<br>
>&emsp;&emsp;&emsp;&ensp;/ &ensp;\ &emsp;&emsp;/ &ensp;\ <br>
>&emsp;&emsp;&ensp;21&ensp;&ensp;22&emsp;23&ensp;&ensp;24<br>
>**Output :**<br>
>10 28<br>
>10 13 15 <br> 
>**Explanation :**<br>
>Sum of path 10, 28 = 38 and<br>
>Sum of path 10, 13, 15 = 38.

<br>

**Your task :**

You don't have to read input or print anything. Your task is to complete the function **printPaths()** that takes the root of the tree and sum as input and returns a vector of vectors containing the paths that lead to the sum.
 
**Expected Time Complexity:** O(N)<br>
**Expected Space Complexity:** O(N<sup>2</sup>)
 
**Your Task :**
- >1 <= N <= 2*10<sup>3</sup>
- >-10<sup>3</sup> <= sum, Node.key <= 10<sup>3</sup>

```python
'''
# Tree Node
class Node:
    def __init__(self, val):
        self.right = None
        self.data = val
        self.left = None
        
'''

class Solution:
    def printPaths(self, root, sum):
        #code here
        result = []

        def dfs(node, current_path, current_sum):
            if not node:
                return

            current_path.append(node.data)
            current_sum += node.data

            if current_sum == sum:
                result.append(current_path.copy())
            
            dfs(node.left, current_path, current_sum)
            dfs(node.right, current_path,current_sum)

            current_path.pop()

        dfs(root, [], 0)
        return result
```
