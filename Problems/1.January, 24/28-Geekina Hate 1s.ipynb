**Your Task:**

This is a function problem. The input is already taken care of by the driver code. You only need to complete the function **findNthNumer()** that takes n and k as input parameters. Return the n<sup>th</sup> number having at most k set-bits.

**Expected Time Complexity:** O(k* log(n) + constant)<br>
**Expected Auxiliary Space:** O(k* log(n) + constant)

**Constraints:**
- >1 <= n <= 10<sup>9</sup>
- >1 <= k <= 63

```python
class Solution:
    def findNthNumber(self, n : int, k : int ) -> int:
        # code here
        from math import factorial, log2

        def combination(n, r):
            return factorial(n) // (factorial(r) * factorial(n - r)) if n >= r else 0
        
        def countset(y, m):
            if m == 0:
                return 1
            if y == 0:
                return 0
            h = int(log2(y))
            
            if h < m - 1:
                return 0
            return combination(h, m) + countset(y^(1 << h), m - 1)
        
        def total(z, p):
            return sum(countset(z, i) for i in range(p + 1))
            
        lower, upper = 0, 10 ** 15
        
        while lower < upper:
            mid = lower + (upper - lower) // 2
            if total(mid, k) >= n:
                upper = mid
            else:
                lower = mid + 1
        
        return lower
```
