Given two linked lists of size **N** and **M**, which are sorted in **non-decreasing** order. The task is to merge them in such a way that the resulting list is in **non-increasing** order.

<br>

**Example 1:**

>**Input:**<br>
>N = 2, M = 2<br>
>list1 = 1->3<br>
>list2 = 2->4<br>
>**Output:**<br>
>4->3->2->1<br>
>**Explanation:**<br>
>After merging the two lists in non-increasing order, we have new lists as 4->3->2->1.

**Example 2:**

>**Input:**<br>
>N = 4, M = 3<br>
>list1 = 5->10->15->40 <br>
>list2 = 2->3->20<br>
>**Output:**
>40->20->15->10->5->3->2<br>
>**Explanation:**<br>
>After merging the two lists in non-increasing order, we have new lists as 40->20->15->10->5->3->2.

<br>

**Your Task:**

The task is to complete the function **mergeResult()** which takes reference to the heads of both linked list and returns the pointer to the merged linked list.

**Expected Time Complexity :** O(N+M)<br>
**Expected Auxiliary Space :** O(1)

**Constraints:**
- >0 <= N, M <= 10<sup>4</sup>

```python
'''
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
'''

class Solution:
    def reverse(self, head):
        curr = head
        pre = None
        
        while(curr != None):
            next = curr.next
            curr.next = pre
            pre = curr
            curr = next
        
        return pre
        
    def mergeResult(self, h1, h2):
        #return head of merged list
        h1 = self.reverse(h1)
        h2 = self.reverse(h2)
        
        # create a new node
        temp = Node(0)
        ans = temp
        
        while h1 and h2:
            if h1.data > h2.data:
                ans.next = h1
                h1 = h1.next
            else:
                ans.next = h2
                h2 = h2.next
            ans = ans.next
        
        while h1:
            ans.next = h1
            h1 = h1.next
            ans = ans.next
        while h2:
            ans.next = h2
            h2 = h2.next
            ans = ans.next
        
        return temp.next
```
