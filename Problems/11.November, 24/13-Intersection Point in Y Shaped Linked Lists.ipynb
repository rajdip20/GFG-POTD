Given two singly linked lists, return the point where two linked lists intersect.

Note: If the linked lists do not merge at any point, return **-1**.

<br>

**Examples:**
>**Input:** Linked list 1: 4->4->4->4->4, Linked list 2: 4->4->4<br>
>![image1](../../images/blobid1_1721764552.png)<br>
>**Output:** 4<br>
>**Explanation:** From the above image, it is clearly seen that the common part is 4->4 whose starting point is 4.

>**Input:** Linked list 1: 4->1->8->4->5, Linked List 2: 5->6->1->8->4->5<br>
>![image2](../../images/blobid0_1721764467.png)<br>
>**Output:** 8<br>
>**Explanation:** From the above image, it is clearly seen that the common part is 8->4->5 whose starting point is 8.

>**Input:** Linked list 1: 1->2->3, Linked list 2: 4->5->6<br>
>**Output:** -1<br>
>**Explanation:** There is no common part, so there is no interaction point.

Try to solve the problem without using any extra space.

<br>

**Constraints:**
- >2 ≤ size of first linkedist + size of second linkedlist ≤ 2*10<sup>5</sup>
- >-10000 ≤ data of nodes ≤ 10000

```python
'''
	Function to return the value at point of intersection
	in two linked list, connected in y shaped form.
	
	Function Arguments: head_a, head_b (heads of both the lists)
	
	Return Type: value in NODE present at the point of intersection
	             or -1 if no common point.

	Contributed By: Nagendra Jha

	{
		# Node Class
		class Node:
		    def __init__(self, data):   # data -> value stored in node
		        self.data = data
		        self.next = None
	}
'''

# Function to find intersection point in Y shaped Linked Lists.
def intersetPoint(head1, head2):
    # code here
    def get_length(head):
        length = 0
        current = head
        while current:
            length += 1
            current = current.next
        return length
    
    lenA = get_length(head1)
    lenB = get_length(head2)
    
    currentA, currentB = head1, head2
    if lenA > lenB:
        for _ in range(lenA - lenB):
            currentA = currentA.next
    else:
        for _ in range(lenB - lenA):
            currentB = currentB.next
    
    while currentA and currentB:
        if currentA == currentB:
            return currentA.data
        currentA = currentA.next
        currentB = currentB.next
    
    return -1
```
