Given a binary tree with **n** nodes and a positive integer **k**, the task is to count the number of special nodes. A node is considered special if there exists at least one leaf in its subtree such that the distance between the node and leaf is exactly k.

**Note:** Any such node should be **counted only once**. For example, if a node is at a distance k from 2 or more leaf nodes, then it would add only 1 to our count.

<br>

**Example 1:**
>**Input:**<br>
>Binary tree<br>
>&emsp;&emsp;&emsp;&emsp; 1 <br>
>&emsp;&emsp;&ensp;&emsp; / &ensp; \ <br>
>&ensp;&ensp;&ensp;&emsp; 2 &emsp;&emsp; 3 <br>
>&emsp;&ensp; / &ensp; \ &emsp; / &ensp; \ <br>
>&emsp; 4 &emsp; 5 &ensp; 6 &emsp; 7 <br>
>&emsp;&emsp;&emsp;&emsp;&emsp; \ <br>
>&emsp;&emsp;&emsp;&emsp;&emsp;&ensp; 8 <br>
>k = 2<br>
>**Output:** <br>
>2<br>
>**Explanation:** <br>
>There are only two unique nodes that are at a distance of 2 units from the leaf node. (node 3 for leaf with value 8 and node 1 for leaves with values 4, 5 and 7) Note that node 2 isn't considered for leaf with value 8 because it isn't a direct ancestor of node 8.


**Example 2:**
>**Input:**<br>
>Binary tree<br>
>&emsp;&emsp;&emsp;&emsp;&emsp; 1 <br>
>&emsp;&emsp;&emsp;&emsp;&ensp; / <br>
>&emsp;&emsp;&emsp;&emsp; 3 <br>
>&emsp;&emsp;&emsp;&ensp; / <br>
>&emsp;&emsp;&emsp; 5 <br>
>&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp; 7 &emsp; 8 <br>
>&emsp;&emsp;&emsp;&emsp;&ensp; \ <br>
>&emsp;&emsp;&emsp;&emsp;&emsp; 9 <br>
>k = 4<br>
>**Output:** <br>
>1<br>
>**Explanation:** <br>
>Only one node is there which is at a distance of 4 units from the leaf node.(node 1 for leaf with value 9) 

<br>

**Your Task:**
You don't have to read input or print anything. Complete the function **printKDistantfromLeaf()** that takes root **node** and **k** as inputs and **returns** the number of nodes that are at distance k from a leaf node. 

**Expected Time Complexity:** O(n).<br>
**Expected Auxiliary Space:** O(Height of the Tree).

**Constraints:**
- >1 <= n <= 10<sup>5</sup>

```python
'''
@input - 
node - root node of given tree
k = distance of nodes required 

class Node:
    def __init__(self, val):
        self.right = None
        self.data = val
        self.left = None
'''

class Solution:
    #Function to return count of nodes at a given distance from leaf nodes.
    def printKDistantfromLeaf(self, root, k):
        #code here
        leaves = set()
        parent = {root: None}
        
        def dfs(node):
            if not node:
                return 
            
            if not node.left and not node.right:
                leaves.add(node)
                return
            
            if node.left:
                parent[node.left] = node
            
            if node.right:
                parent[node.right] = node
            
            dfs(node.left)
            dfs(node.right)
        
        dfs(root)
        visited = set()
        answer = 0
        
        for e in leaves:
            node = 0
            
            while node < k and e:
                node += 1
                e = parent[e]
            
            if e and e not in visited:
                visited.add(e)
                answer += 1
        
        return answer
```
