 Consider Red lines of slope -1 passing between nodes (in following diagram). The diagonal sum in a binary tree is the sum of all node datas lying between these lines. Given a Binary Tree of size **n**, print all diagonal sums.

For the following input tree, output should be 9, 19, 42.<br>
9 is sum of 1, 3 and 5.<br>
19 is sum of 2, 6, 4 and 7.<br>
42 is sum of 9, 10, 11 and 12.



<br>

**Example 1:**
>**Input:** <br>
>&emsp;&emsp;&emsp;&emsp; 4 <br>
>&emsp;&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp;&emsp; 1 &emsp; 3 <br>
>&emsp;&emsp;&emsp;&emsp;&ensp; / <br>
>&emsp;&emsp;&emsp;&emsp; 3 <br>
>**Output:** <br>
>7 4 

**Example 2:**
>**Input:** <br>
>&emsp;&emsp;&emsp;&emsp; 10 <br>
>&emsp;&emsp;&emsp;&ensp; / &emsp; \ <br>
>&emsp;&emsp;&emsp; 8 &emsp;&ensp; 2 <br>
>&emsp;&emsp;&ensp; / &nbsp; \ &ensp; / <br>
>&emsp;&emsp; 3 &ensp;&nbsp; 5  2 <br>
>**Output:** <br>
>12 15 3 

<br>

**Your Task:**<br>
You don't need to take input. Just complete the function **diagonalSum()** that takes root **node** of the tree as parameter and returns an array containing the diagonal sums for every diagonal present in the tree with slope -1.

**Expected Time Complexity:** O(nlogn).<br>
**Expected Auxiliary Space:** O(n).

**Constraints:**
- >1 <= n <= 10<sup>5</sup>
- >0 <= data of each node <= 10<sup>4</sup>

```python
'''
# Node Class:
class Node:
    def __init__(self,val):
        self.data = val
        self.left = None
        self.right = None
'''

class Solution:
    # Complete the function below
    def diagonalSum(self, root):
        # :param root: root of the given tree.
        
        # code here
        from collections import deque

        
        count = {}
        nodes = deque()
        nodes.append((root,0))
        
        while nodes:
            node, level = nodes.popleft()
            
            if level not in count:
                count[level] = 0
            
            count[level] += node.data
            if node.left:
                nodes.append((node.left, level + 1))
            
            if node.right:
                nodes.append((node.right, level))
        
        result = [0] * len(count)
        for i in range(len(count)):
            result[i] = count[i]
        
        return result
```
