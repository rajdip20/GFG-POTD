Given the **root of a binary search tree** and a number **n**, find the greatest number in the binary search tree that is less than or equal to **n**. 

<br>

**Example 1 :**
>**Input :** <br>
><br>
><br>
>n = 24<br>
>**Output :** <br>
>21<br>
>**Explanation :** The greatest element in the tree which is less than or equal to 24, is 21. (Searching will be like 5->12->21)

**Example 2 :**
>**Input :** <br>
><br>
><br>
>n = 4<br>
>**Output :** <br>
>3<br>
>**Explanation :** The greatest element in the tree which is less than or equal to 4, is 3. (Searching will be like 5->2->3)

<br>

**Your task :**<br>
You don't need to read input or print anything. Your task is to complete the function **findMaxForN()** which takes the **root** of the BST, and the integer **n** as input parameters and returns the **greatest element less than or equal** to **n** in the given BST, Return -1 if no such element exists.

**Expected Time Complexity:** O(Height of the BST)<br>
**Expected Auxiliary Space:** O(Height of the BST).

**Constraints:**
- >1 <= n <= 10<sup>3</sup>

```python
'''
class Node:
    """ Class Node """
    def __init__(self, value):
        self.left = None
        self.key = value
        self.right = None
'''

class Solution:
    def findMaxForN(self, root, n):
        # Print the value of the element if it exists otherwise print -1.
        
        # code here
        result = -1
    
        while root:
            if root.key <= n:
                result = root.key
                root = root.right
            else:
                root = root.left
                
        return result
```
