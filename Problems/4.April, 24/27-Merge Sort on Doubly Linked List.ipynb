Given Pointer/Reference to the **head** of a **doubly linked list** of **n** nodes, the task is to **Sort** the **given doubly linked** list using **Merge Sort** in both **non-decreasing** and **non-increasing** order.

<br>

**Example 1:**
>**Input:**<br>
>n = 8<br>
>value[] = {7,3,5,2,6,4,1,8}<br>
>**Output:**<br>
>1 2 3 4 5 6 7 8<br>
>8 7 6 5 4 3 2 1<br>
>**Explanation:**<br>
>After sorting the given linked list in both ways, resultant matrix will be as given in the first two line of output, where first line is the output for non-decreasing order and next line is for non-increasing order.

**Example 2:**
>**Input:**<br>
>n = 5<br>
>value[] = {9,15,0,-1,0}<br>
>**Output:**<br>
>-1 0 0 9 15<br>
>15 9 0 0 -1<br>
>**Explanation:**<br>
>After sorting the given linked list in both ways, the resultant list will be -1 0 0 9 15 in non-decreasing order and 15 9 0 0 -1 in non-increasing order.

<br>

**Your Task:**<br>
The task is to complete the function **sortDoubly()** which takes reference to the **head** of the doubly linked and **Sort** the given doubly linked list using **Merge Sort** in both **non-decreasing** and **non-increasing**. The **printing** is done **automatically** by the **driver code**.

**Expected Time Complexity:** O(nlogn)<br>
**Expected Space Complexity:** O(logn)

**Constraints:**
- >1 <= n <= 10<sup>4</sup>
- >-10<sup>5</sup> <= values[i] <= 10<sup>5</sup>

```python
'''
class Node:
	def __init__(self, data):
		self.data = data
		self.next = None
		self.prev = None
'''

class Solution():
    # Function to sort the given doubly linked list using Merge Sort.
    def sortDoubly(self, head):
        # Your code here
        def merge_sort(start=head):
            if not start.next:
                return start
            
            slow, fast = start, start
            while fast and fast.next:
                slow = slow.next
                fast = fast.next.next
            
            mid = slow
            mid.prev.next = None
            mid.prev = None
            left = merge_sort(start)
            right = merge_sort(mid)
            
            dummy = Node(None)
            current = dummy
            
            while left or right:
                if left and right:
                    if left.data <= right.data:
                        current.next = left
                        left.prev = current
                        left = left.next
                    else:
                        current.next = right
                        right.prev = current
                        right = right.next
                elif left:
                    current.next = left
                    left.prev = current
                    left = None
                elif right:
                    current.next = right
                    right.prev = current
                    right = None
                current = current.next
            
            dummy.next.prev = None
            
            return dummy.next

        return merge_sort()
```
