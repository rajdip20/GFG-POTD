Given a BST with **n (n>=2)** nodes, find the **kth** common ancestor of nodes **x** and **y** in the given tree. Return **-1** if kth ancestor does not exist.

Nodes x and y will always be **present** in the input of a BST, and **x != y**.

<br>

**Example 1:**
>**Input:** <br>
>Input tree<br>
><br>
>k = 2, x = 40, y = 60 <br>
>**Output:**<br>
>30<br>
>**Explanation:**<br>
>Their 2nd common ancestor is 30.

**Example 2:**
>**Input:** <br>
>Input tree<br>
><br>
>k = 2, x = 40, y = 60<br>
>**Output:**<br>
>-1<br>
>**Explanation:**<br>
>LCA of 40 and 60 is 50, which is root itself. There does not exists 2nd common ancestor in this case.

<br>

**Your task :**<br>
You don't have to read input or print anything. Your task is to complete the function **kthCommonAncestor()** that takes the **root** of the tree, **k, x** and **y** as input and returns the kth common ancestor of x and y.
 
**Expected Time Complexity:** O(Height of the Tree)<br>
**Expected Space Complexity:** O(Height of the Tree)
 
**Your Task :**
- >1 <= n, k <= 10<sup>5</sup>
- >1 <= node->data, x, y <= 10<sup>9</sup>

```python
'''
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
'''

class Solution:
    def kthCommonAncestor(self, root, k, x, y):
        # Code here
        xn, yn = None, None
        parent = {}
        
        def dfs(n, p = None):
            nonlocal xn, yn
            parent[n] = p
            
            if n.data == x:
                xn = n
            if n.data == y:
                yn = n
            if n.left:
                dfs(n.left, n)
            if n.right:
                dfs(n.right, n)
        
        dfs(root)
        ps = set()
        
        while xn:
            ps.add(xn)
            xn = parent[xn]
        
        while yn:
            if yn in ps:
                break
            yn = parent[yn]
   
        while yn and k > 1:
            k -= 1
            yn = parent[yn]
        return yn.data if yn else -1
```
