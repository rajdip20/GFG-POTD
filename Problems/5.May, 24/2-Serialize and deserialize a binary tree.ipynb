Serialization is to store a tree in an array so that it can be later restored and deserialization is reading tree back from the array. Complete the functions
- **serialize() :** stores the tree into an array a and returns the array.
- **deSerialize() :** deserializes the array to the tree and returns the root of the tree.

**Note:** Multiple nodes can have the same data and the node values are **always positive integers**. Your code will be correct if the tree returned by **deSerialize(serialize(input_tree))** is same as the input tree. Driver code will print the in-order traversal of the tree returned by deSerialize(serialize(input_tree)).

<br>

**Example 1:**
>**Input:**<br>
>&emsp;&emsp;&emsp; 1 <br>
>&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp; 2 &emsp; 3 <br>
>**Output:** <br>
>2 1 3

**Example 2:**
>**Input:**<br>
>&emsp;&emsp;&emsp;&emsp;&emsp; 10 <br>
>&emsp;&emsp;&emsp;&emsp;&ensp; / &emsp; \ <br>
>&emsp;&emsp;&emsp;&ensp; 20 &emsp;&ensp; 30 <br>
>&emsp;&emsp;&emsp; / &ensp; \ <br>
>&emsp;&emsp; 40 &emsp; 60 <br>
>**Output:** <br>
>40 20 60 10 30

**Your Task:**<br>
The task is to complete two functions **serialize** which takes the root node of the tree as input and stores the tree into an array and **deSerialize** which deserializes the array to the original tree and returns the root of it.

**Expected Time Complexity:** O(Number of nodes).<br>
**Expected Auxiliary Space:** O(Number of nodes).

**Constraints:**
- >1 <= Number of nodes <= 10<sup>4</sup>
- >1 <= Data of a node <= 10<sup>9</sup>

```python
'''
class Node:
    def __init__(self,val):
        self.data = val
        self.left = None
        self.right = None
'''

class Solution:
    # Function to serialize a tree and return a list containing nodes of tree.
    def serialize(self, root):
        # code here
        if not root:
            return ['N']
        
        queue = [root]
        result = []
        
        while queue:
            node = queue.pop(0)
            if node:
                result.append(str(node.data))
                queue.append(node.left)
                queue.append(node.right)
            else:
                result.append('N')
        
        return result
    
    # Function to deserialize a list and construct the tree.   
    def deSerialize(self, a):
        # code here
        if not a or a[0] == 'N':
            return None
        
        root = Node(int(a.pop(0)))
        queue = [root]
        
        while a:
            node = queue.pop(0)
            left_val = a.pop(0)
            if left_val != 'N':
                node.left = Node(int(left_val))
                queue.append(node.left)
            
            if a:
                right_val = a.pop(0)
                if right_val != 'N':
                    node.right = Node(int(right_val))
                    queue.append(node.right)
        
        return root
```
