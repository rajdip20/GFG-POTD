Given a sorted array of unique elements in increasing order, **arr[]** of **n** integers, and a value **x**. Find the K closest elements to **x** in **arr[]**.<br>
Keep the following points in mind:
1. If **x** is present in the array, then it need not be considered.
2. If two elements have the same difference as **x**, the greater element is prioritized.
3. If sufficient elements are not present on the right side, take elements from the left and vice versa.

<br>

**Example 1:**
>**Input:**<br>
>n = 13<br>
>arr[] = {12, 16, 22, 30, 35, 39, 42, 45, 48, 50, 53, 55, 56}<br>
>k = 4, x = 35<br>
>**Output:** 39 30 42 45<br>
>**Explanation:** <br>
>First closest element to 35 is 39.<br>
>Second closest element to 35 is 30.<br>
>Third closest element to 35 is 42.<br>
>And fourth closest element to 35 is 45.

**Example 2:**
>**Input:**<br>
>n = 5<br>
>arr[] = {1, 2, 3, 6, 10}<br>
>k = 3, x = 4<br>
>**Output:** 3 6 2<br>
>**Explanation:** <br>
>First closest element is 3.<br>
>There are two elements 2 and 6 for which <br>
>the difference with 4 is same i.e. 2.<br>
>So first take greatest number 6 <br>
>then the lower number 2.

<br>

**Your Task:**<br>
You don't need to read input or print anything. Complete the function **printKClosest()** which takes **arr[]**, **n**, **k**, and **x** as input parameters and returns an array of integers containing the K closest elements to **x** in **arr[]**.


**Expected Time Complexity:** O(log n + k)<br>
**Expected Auxiliary Space:** O(k)


**Constraints:**
- >1 ≤ n ≤ 10<sup>5</sup>
- >1 ≤ k ≤ n
- >1 ≤ x ≤ 10<sup>6</sup>
- >1 ≤ arr[i] ≤ 10<sup>6</sup>

```python
class Solution:
    def printKClosest(self, arr, n, k, x):
        # code here
        from heapq import heappush, heappop


        heap = []
        
        for val in arr[::-1] :
            if val==x:
                continue
            dist = abs(val - x)
            
            if len(heap) < k:
                heappush(heap, (-1 * dist, val))
            else:
                if -1 * heap[0][0] > dist:
                    heappop(heap)
                    heappush(heap, (-1 * dist, val))
                    
        result = []
        while heap:
            _, val = heappop(heap)
            result.append(val)
        return result[::-1]
```
