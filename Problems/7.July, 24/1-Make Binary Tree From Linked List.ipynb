Given a Linked List Representation of Complete Binary Tree. The task is to construct the Binary tree and print the level order traversal of the Binary tree.<br>
**Note:** The complete binary tree is represented as a linked list in a way where if the root node is stored at position i, its left, and right children are stored at position **2 * i+1**, and **2 * i+2** respectively. H is the height of the tree and this space is used implicitly for the recursion stack.

<br>

**Examples:**
>**Input:** n = 5, k = 1->2->3->4->5 <br>
>**Output:** 1 2 3 4 5 <br>
>**Explanation:** The tree would look like <br>
>&emsp;&emsp; &emsp; 1 <br>
>&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp; 2 &emsp; 3 <br>
>&emsp;&ensp; / &ensp; \ <br>
>&emsp; 4 &emsp; 5 <br>
>Now, the level order traversal of <br>
>the above tree is 1 2 3 4 5.

>**Input:** n = 5, k = 5->4->3->2->1 <br>
>**Output:** 5 4 3 2 1 <br>
>**Explanation:** The tree would look like <br>
>&emsp;&emsp;&emsp; 5 <br>
>&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp; 4 &emsp; 3 <br>
>&emsp;&ensp; / &ensp; \ <br>
>&emsp; 2 &emsp; 1 <br>
>Now, the level order traversal of <br>
>the above tree is 5 4 3 2 1.

<br>

**Expected Time Complexity:** O(n). <br>
**Expected Auxiliary Space:** O(n).

**Constraints:**
- >1 <= n <= 10<sup>5</sup>
- >1 <= k<sub>i</sub> <= 10<sup>5</sup>

```python
'''
class ListNode:

    # Constructor to create a new node
    def __init__(self, data):
        self.data = data
        self.next = None


# Tree Node structure
class Tree:

    # Constructor to create a new node
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

'''

# Function to make binary tree from linked list.
def convert(head):
    # code here
    from collections import deque

    if head.next == None:
        return Tree(head.data)
    
    que = deque()
    root = Tree(head.data)
    que.append(root)
    head = head.next
    
    while head:
        current = que.popleft()
        current.left = Tree(head.data)
        que.append(current.left)
        head = head.next
        if head:
            current.right = Tree(head.data)
            que.append(current.right)
            head = head.next
    
    return root
```
