Given two BSTs, return elements of merged BSTs in **sorted** form.

 <br>

>**Examples:**
>**Input:** <br>
>BST1: <br>
>&emsp;&emsp;&emsp;&emsp; 5 <br>
>&emsp;&emsp;&emsp;&ensp; / &nbsp; \ <br>
>&emsp;&emsp;&emsp; 3 &emsp; 6 <br>
>&emsp;&emsp;&nbsp; / &ensp; \ <br>
>&emsp;&ensp;&nbsp; 2 &emsp; 4   <br>
>BST2: <br>
>&emsp;&emsp;&emsp; 2 <br>
>&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp; 1 &emsp; 3 <br>
>&emsp;&emsp;&emsp;&emsp;&ensp; \ <br>
>&emsp;&emsp;&emsp;&emsp;&emsp; 7 <br>
>&emsp;&emsp;&emsp;&emsp;&ensp; / <br>
>&emsp;&emsp;&emsp;&emsp; 6 <br>
>**Output:** 1 2 2 3 3 4 5 6 6 7 <br>
>**Explanation:** After merging and sorting the two BST we get 1 2 2 3 3 4 5 6 6 7.

>**Input:**<br>
>BST1:<br>
>&emsp;&emsp;&emsp;&emsp; 12<br>
>&emsp;&emsp;&emsp;&ensp; /   <br>
>&emsp;&emsp;&emsp; 9<br>
>&emsp;&emsp;&nbsp; / &ensp; \ <br>
>&emsp;&ensp;&nbsp; 6 &emsp; 11<br>
>BST2:<br>
>&emsp;&emsp;&emsp;&emsp; 8 <br>
>&emsp;&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp;&emsp; 5 &emsp; 10 <br>
>&emsp;&emsp;&nbsp; / <br>
>&emsp;&ensp; 2 <br>
>**Output:** 2 5 6 8 9 10 11 12<br>
>**Explanation:** After merging and sorting the two BST we get 2 5 6 8 9 10 11 12.

<br>

**Expected Time Complexity:** O(m+n)<br>
**Expected Auxiliary Space:** O(Height of BST1 + Height of BST2 + m + n)

**Constraints:**
- >1 ≤ Number of Nodes ≤ 10<sup>5</sup>

```python
class Solution:
    def merge(self, root1, root2):
        # code here
        def inorder_traversal(root):
            if not root:
                return []
            return inorder_traversal(root.left) + [root.data] + inorder_traversal(root.right)


        def merge_two_sorted_lists(list1, list2):
            merged_list = []
            i, j = 0, 0
            while i < len(list1) and j < len(list2):
                if list1[i] < list2[j]:
                    merged_list.append(list1[i])
                    i += 1
                else:
                    merged_list.append(list2[j])
                    j += 1

            while i < len(list1):
                merged_list.append(list1[i])
                i += 1

            while j < len(list2):
                merged_list.append(list2[j])
                j += 1
            return merged_list
        
        list1 = inorder_traversal(root1)
        list2 = inorder_traversal(root2)

        return merge_two_sorted_lists(list1, list2)
```
