Given a binary tree and a **node data** called **target**. Find the minimum time required to burn the complete binary tree if the target is set on fire. It is known that in 1 second all nodes connected to a given node get burned. That is its left child, right child, and parent.<br>
**Note:** The tree contains unique values.

<br>

**Examples:**
>**Input:** <br>
>&emsp;&emsp;&emsp;&emsp;&emsp; 1 <br>
>&emsp;&emsp;&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp;&emsp;&emsp; 2 &emsp;&nbsp; 3 <br>
>&emsp;&emsp;&emsp;&ensp; / &ensp; \ &emsp; \ <br>
>&emsp;&emsp;&emsp; 4 &emsp; 5 &emsp; 6 <br>
>&emsp;&emsp;&emsp;&emsp;&ensp; / &ensp; \ &emsp; \ <br>
>&emsp;&emsp;&emsp;&emsp; 7 &emsp; 8 &emsp; 9 <br>
>&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&ensp; \ <br>
>&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&ensp; 10 <br>
>Target Node = 8 <br>
>**Output:** 7 <br>
>**Explanation:** If leaf with the value 8 is set on fire. <br>
>After 1 sec: 5 is set on fire. <br>
>After 2 sec: 2, 7 are set to fire. <br>
>After 3 sec: 4, 1 are set to fire. <br>
>After 4 sec: 3 is set to fire. <br>
>After 5 sec: 6 is set to fire. <br>
>After 6 sec: 9 is set to fire. <br>
>After 7 sec: 10 is set to fire. <br>
>It takes 7s to burn the complete tree.

>**Input:** <br>
>&emsp;&emsp;&emsp;&emsp; 1 <br>
>&emsp;&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp;&emsp; 2 &emsp;&nbsp; 3 <br>
>&emsp;&emsp;&ensp; / &ensp; \ &emsp; \ <br>
>&emsp;&emsp; 4 &emsp; 5 &emsp; 7 <br>
>&emsp;&ensp; / &emsp;&ensp; / <br>
>&emsp; 8 &emsp; 10 <br>
>Target Node = 10 <br>
>**Output:** 5

<br>

**Expected Time Complexity:** O(number of nodes) <br>
**Expected Auxiliary Space:** O(height of tree)

<br>

**Constraints:**
- >1 ≤ number of nodes ≤ 10<sup>5</sup>
- >1 ≤ values of nodes ≤ 10<sup>5</sup>

```python
class Solution:
    def minTime(self, root, target):
        # code here
        from collections import deque

        if not root:
            return 0
        
        parent_map = {}
        target_node = None
        
        def dfs(node, parent):
            nonlocal target_node
            if not node:
                return
            if node.data == target:
                target_node = node
            parent_map[node] = parent
            dfs(node.left, node)
            dfs(node.right, node)
        
        dfs(root, None)
        
        time_to_burn = 0
        visited = set()
        queue = deque([target_node])
        visited.add(target_node)
        
        while queue:
            size = len(queue)
            for _ in range(size):
                current_node = queue.popleft()
                
                if current_node.left and current_node.left not in visited:
                    queue.append(current_node.left)
                    visited.add(current_node.left)
                    
                if current_node.right and current_node.right not in visited:
                    queue.append(current_node.right)
                    visited.add(current_node.right)
                    
                parent_node = parent_map.get(current_node)
                if parent_node and parent_node not in visited:
                    queue.append(parent_node)
                    visited.add(parent_node)
            
            if queue:
                time_to_burn += 1
        
        return time_to_burn
```
