Given two BSTs, return elements of merged BSTs in **sorted** form.

<br>

**Examples:**
>**Input:** <br>
>BST1: <br>
>&emsp;&emsp;&emsp; 5 <br>
>&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp; 3 &emsp; 6 <br>
>&emsp;&ensp; / &ensp; \ <br>
>&emsp; 2 &emsp; 4 <br>
>BST2: <br>
>&emsp;&emsp; 2 <br>
>&emsp;&ensp; / &ensp; \ <br>
>&emsp; 1 &emsp; 3 <br>
>&emsp;&emsp;&emsp;&ensp; \ <br>
>&emsp;&emsp;&emsp;&emsp; 7 <br>
>&emsp;&emsp;&emsp;&ensp; / <br>
>&emsp;&emsp;&emsp; 6 <br>
>**Output:** 1 2 2 3 3 4 5 6 6 7 <br>
>**Explanation:** After merging and sorting the two BST we get 1 2 2 3 3 4 5 6 6 7.

>**Input:** <br>
>BST1: <br>
>&emsp;&emsp;&emsp; 12 <br>
>&emsp;&emsp;&ensp; / <br>
>&emsp;&emsp; 9 <br>
>&emsp;&ensp; / &ensp; \ <br>
>&emsp; 6 &emsp; 11 <br>
>BST2: <br>
>&emsp;&emsp;&emsp; 8 <br>
>&emsp;&emsp;&ensp; / &ensp; \ <br>
>&emsp;&emsp; 5 &emsp; 10 <br>
>&emsp;&ensp; / <br>
>&emsp; 2 <br>
>**Output:** 2 5 6 8 9 10 11 12 <br>
>**Explanation:** After merging and sorting the two BST we get 2 5 6 8 9 10 11 12.

<br>

**Expected Time Complexity:** O((m+n) * log(m+n))<br>
**Expected Auxiliary Space:** O(Height of BST1 + Height of BST2 + m + n)

**Constraints:**
- >1 ≤ Number of Nodes, value of nodes ≤ 10<sup>5</sup>

```python
class Solution:
    def merge(self, root1, root2):
        # code here
        def inorder_traversal(root, result):
            if not root:
                return
            inorder_traversal(root.left, result)
            result.append(root.data)
            inorder_traversal(root.right, result)

        def merge_sorted_arrays(arr1, arr2):
            i, j = 0, 0
            merged = []
            
            while i < len(arr1) and j < len(arr2):
                if arr1[i] <= arr2[j]:
                    merged.append(arr1[i])
                    i += 1
                else:
                    merged.append(arr2[j])
                    j += 1
            
            while i < len(arr1):
                merged.append(arr1[i])
                i += 1
            
            while j < len(arr2):
                merged.append(arr2[j])
                j += 1
            
            return merged

        bst1_elements = []
        bst2_elements = []
        
        inorder_traversal(root1, bst1_elements)
        inorder_traversal(root2, bst2_elements)
        
        return merge_sorted_arrays(bst1_elements, bst2_elements)
```
